```python
import random
xMin1 = 1
xMax1 = 2
yMin1 = 2
yMax1 = 3
xMin2 = 1.5
xMax2 = 2.5
yMin2 = 2.5
yMax2 = 3.5
x = []
y = []
for i in range(50):
    a = [random.uniform(xMin1, xMax1),random.uniform(yMin1, yMax1)]
    x.append(a)
    y.append(0)
for i in range(50):
    a = [random.uniform(xMin2, xMax2),random.uniform(yMin2, yMax2)]
    x.append(a)
    y.append(1)
print(x)
print(y)
```

```python
import matplotlib.pyplot as plt
import numpy as np
xx=[]
xy=[]
for i in range(len(x)):
    xx.append(x[i][0])
for i in range(len(x)):
    xy.append(x[i][1])
fig, ax = plt.subplots() 
# Create a figure containing a single Axes.
ax.scatter(xx[:50],xy[:50],c='r')  # Plot some data on the Axes.
ax.scatter(xx[50:],xy[50:],c='g') 
plt.show()                           # Show the figure.
```

```python
x_test = [] # 20 значений хранятся в этом списке
x_train = [] # 80 значений
y_test = [0]*10 + [1]*10
y_train = [0]*40 + [1]*40
def train_test_split(x,y):
    for i in range(10):
        x_test.append(x[i])
    for i in range(90,100):
        x_test.append(x[i])
    for i in range(10,90):
        x_train.append(x[i])
train_test_split(x,y)
print(x_test)
print(x_train)
print(y_test)
print(y_train)
```

```python
from heapq import nsmallest
y_predict = []
def fit(x_train,y_train,x_test): # Нужно вывести y_test (данные для тестирования)
    for i in range(len(x_test)):
        distance=[]
        for j in range(len(x_train)):
            distance.append(((x_train[j][0]-x_test[i][0])**2+(x_train[j][1]-x_test[i][1])**2)**0.5)
        a, b,c = map(distance.index, nsmallest(3, distance))
        list = [a,b,c]
        s=0
        for i in list:
            if i>=40: s+=1
            else: s+=0
        if s>1: y_predict.append(1)
        else: y_predict.append(0)
fit(x_train,y_train,x_test)
print(y_predict)
```

```python
def computeAccuracy (y_test,y_predict):
    KC=0
    N = 20
    for i in range(N):
        if y_predict[i]== y_test[i]:
            KC+=1
    accuracy = KC/N
    print (accuracy)
computeAccuracy (y_test,y_predict)
```
