```python
# 2, 2 
class People:
    def __init__(self,name,maths,phisics,Russian):
        self.name = name
        self.maths = maths
        self.phisics = phisics
        self.Russian = Russian
        if int(self.maths)==2 or int(self.phisics)==2 or int(self.Russian)==2:
            self.evaluation = 0
        else:
            self.evaluation = (int(self.maths)+int(self.phisics)+int(self.Russian))/3
        
            
name = 1
d={}
while name !='stop':
    name = input(' : ')
    if name == 'stop':
        break
    maths = input('   : ')
    phisics = input('   : ')
    Russian = input('   : ')
    a = People(name,maths,phisics,Russian)
    d.update({a.evaluation:a.name})

for key, value in dict(d).items():
  if key == 0:
    del d[key]

sorted_items = sorted(d.items(), reverse=True)  
sorted_dict = dict(sorted_items)

print('      ,   ')
for key, value in dict(sorted_dict).items():
    print(value)
```

```python
class People:
    def init(self,surename,one,two,three):
        self.surename = surename
        self.one = one
        self.two = two
        self.three = three
        self.m = max(self.one,self.two,self.three)


surename = ''
d = {}
while surename!='stop':
    surename = input(' ')
    if surename == 'stop':
        break
    one = input('  1')
    two = input('  2')
    three = input('  3')
    chelovek = People(surename,one,two,three)
    d.update({chelovek.surename:chelovek.m})
print('     ')
print(d)
```
