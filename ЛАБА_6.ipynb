```python
#Задание 2, 2 уровень
class People:
    def __init__(self,name,maths,phisics,Russian):
        self.name = name
        self.maths = maths
        self.phisics = phisics
        self.Russian = Russian
        if int(self.maths)==2 or int(self.phisics)==2 or int(self.Russian)==2:
            self.evaluation = 0
        else:
            self.evaluation = (int(self.maths)+int(self.phisics)+int(self.Russian))/3
        
            
name = 1
d={}
while name !='stop':
    name = input('Введите имя: ')
    if name == 'stop':
        break
    maths = input('Введите оценку по математике: ')
    phisics = input('Введите оценку по физике: ')
    Russian = input('Введите оценку по русскому: ')
    a = People(name,maths,phisics,Russian)
    d.update({a.evaluation:a.name})

for key, value in dict(d).items():
  if key == 0:
    del d[key]

sorted_items = sorted(d.items(), reverse=True)  
sorted_dict = dict(sorted_items)

print('Вывожу учеников в порядке убывания среднего балла, которых не отчислили')
for key, value in dict(sorted_dict).items():
    print(value)
```
