```python
import random
from random import randint
random.seed(214)
x = randint(1,24)
y = randint(1,28)
print(x) # Получилось задание 1.17 
print(y) # Получилось задание 1.24
```

```python
''' 1.Напишите программу, строящую график функции. Коэффициенты a,b,c и диапазон задаются с клавиатуры.
Проведите проверку решения задачи в Excel.
f(x)=a+ b·x**2+c·ln x'''

import matplotlib.pyplot as plt # Импортируем модуль библиотеки
import numpy as np

x = np.linspace(1, 10, 200) # Первый аргумент является первый элементом списка Создание линейного пространства
#s = str(input('Введите через пробел числа a,b,c'))
s = '1 2 3'
s = s.split(' ')
a = int(s[0])
b = int(s[1])
c = int(s[2])
y = a+b*x**2+c*np.log(x) # Получить значение функции результатом которого хотим получить график

fig, ax = plt.subplots() # Создает основу графика
ax.plot(x, y) # Построение графика по точкам
plt.show() # Вызываем функцию, чтобы вывести все что мы делаем на экран
# График строитмся на основе 200 точек по x и 200 точек по y
```

```python
''' 2. Напишите программу построения графика по имеющемуся дискретному набору известных значений. Для этого:
• Данные по выданному варианту поместите в файл. 
• С помощью программы прочитайте их.
• Постройте график в Python.
• Проведите проверку решения задачи в Excel.
№24.
Х 270 327 627 927 1077 1227 1377 1527 1677 1827 2127
У 406 186 160 140 110      80      65      50      40      30      12
'''
with open ("File.txt","w") as file:
    file.write("[270,406],[327,186],[627,160],[927,140],[1077,110],[1227,80],[1377,65],[1527,50],[1677,40],[1827,30],[2127,12]")#Записали в файл числа
FirstFile = ""
with open("File.txt","r") as file:
    for line in file:
        FirstFile+=line
FirstFile=FirstFile.split(',')
x = []
y = []
for i in FirstFile:
    if i[0]=='[':
        x.append(int(i[1:]))
    elif i[-1]==']':
        y.append(int(i[:-1]))
        
for i in range (1,len(FirstFile)-2):
    if FirstFile[i-1]=='[':
        x.append(int(FirstFile[i]))
    elif FirstFile[i+1]==']':
        y.append(int(FirstFile[i]))
        
for i in range (1,len(FirstFile)-2):
    if FirstFile[i-1]=='[':
        x.append(int(FirstFile[i]))
    elif FirstFile[i+1]==']':
        y.append(int(FirstFile[i]))

fig, ax = plt.subplots() # Создает основу графика
ax.plot(x, y) # Построение графика по точкам
```
