```python
#Поменять местами максимальные элементы матриц А размером 5 на 6 и В размером 3 на 5. Поиск максимального элемента матрицы осуществить в мептоде
def f(A):
    m = int(A[0][0])
    k=[0]*2
    for i in range(len(A)):
        for j in range(len(A[i])):
            if A[i][j]>m:
                m=A[i][j]
                k[0]=i
                k[1]=j
    return k
A = [[1,2,3,4,5,6],
     [1,2,3,8,5,6],
     [1,2,3,4,5,6],
     [1,2,3,4,5,6],
     [1,2,3,4,5,6]]
B = [[1,2,3,4,5],
     [1,2,3,9,5],
     [1,2,3,4,5]]
A[f(A)[0]][f(A)[1]], B[f(B)[0]][f(B)[1]] = B[f(B)[0]][f(B)[1]],A[f(A)[0]][f(A)[1]]
for i in A:
    print(i)
for i in B:
    print(i)
```
