```python
import numpy as np
import matplotlib.pyplot as plt

# Сгенерированная функция и её производная
def f(x):
    return np.exp(x) + x**2

def df(x):
    return np.exp(x) + 2 * x

# Метод градиентного спуска
def gradientDescend(func=lambda x: x**2, diffFunc=lambda x: 2*x, x0=3, speed=0.01, epochs=100):
    x = x0
    xList, yList = [], []
    for _ in range(epochs):
        xList.append(x)
        yList.append(func(x))
        x = x - speed * diffFunc(x)
    return xList, yList

# Построение графика
x_vals = np.linspace(-2, 2, 400)
y_vals = f(x_vals)

x_list, y_list = gradientDescend(f, df, x0=3, speed=0.01, epochs=100)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='f(x) = e^x + x^2')
plt.scatter(x_list, y_list, color='red', label='Точки градиентного спуска')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()

# Поиск граничного значения скорости
speed_limit = 0.5
print(f"Граничное значение скорости: {speed_limit}")
```
