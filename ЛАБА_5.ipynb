```python
import math
import numpy as nm
def f(a,b,h):
    for x in nm.arange(a,b,h):
        sinx = math.sin(x)
        cossinx = math.cos(sinx)
        cosx = math.cos(x)
        ex = math.exp(cosx)
        y = ex*cossinx
        s = 1
        for i in range(1,1000):
            cosix= math.cos(i*x)
            ifact = math.factorial(i)
            if (cosix/ifact)<0.0001:
                break
            s = s+(cosix/ifact)
        print(y,s)
print(f(0.1,1,0.1))
```

```python
import math
import numpy as nm
def f(a,b,h):
    for x in nm.arange(a,b,h):
        pi = math.pi
        x = x*pi
        y = ((x)**2-(pi**2)/3)/4
        s = 1
        for i in range(1,1000):
            cosix = math.cos(i*x)
            s = s+(-1)**i*cosix/(i**2)
            if (-1)**i*cosix/(i**2)<0.0001:
                break
            print(y,s)
print(f(0.2,1,0.04)) #  a,b,c   
```
