```python
import numpy as np
import matplotlib.pyplot as plt #библиотека для графиков
from sklearn.model_selection import train_test_split #инструмент для разделения набора данных на обучающую и тестовую выборки.
from sklearn.neighbors import KNeighborsClassifier #Он предсказывает класс объекта на основе классов его k ближайших соседей в пространстве признаков.
from sklearn.metrics import accuracy_score #то инструмент для вычисления точности (accuracy) классификатора.
from random import uniform

import random
xMin1 = 1
xMax1 = 2
yMin1 = 2
yMax1 = 3
xMin2 = 1.5
xMax2 = 2.5
yMin2 = 2.5
yMax2 = 3.5
x = []
y = []
for i in range(50):
    a = [random.uniform(xMin1, xMax1),random.uniform(yMin1, yMax1)]
    x.append(a)
    y.append(0)
for i in range(50):
    a = [random.uniform(xMin2, xMax2),random.uniform(yMin2, yMax2)]
    x.append(a)
    y.append(1)
x = np.array(x)
y = np.array(y)
print(x)
print(y)

```

```python
# 2. Разделение данных на обучающую и тестовую выборки (используем sklearn)
# Встроенная функция train_test_split из sklearn уже делает это

# 3. Реализация метода k ближайших соседей (используем KNeighborsClassifier из sklearn)

# 4. Расчет метрики accuracy (используем accuracy_score из sklearn)

# Основной код
# Разделение на обучающую и тестовую выборки
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
    
# Создание и обучение модели k ближайших соседей
k = 3
model = KNeighborsClassifier(n_neighbors=k)
model.fit(x_train, y_train)

#Предсказание на тестовых данных
y_pred = model.predict(x_test)
    
# Расчет точности
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
    
# Визуализация результатов
plt.figure(figsize=(10, 8))

# Обучающие точки - класс 0
mask_train0 = y_train == 0
plt.scatter(x_train[mask_train0, 0], x_train[mask_train0, 1], c='blue', marker='o', label='Train Class 0')
 
# Обучающие точки - класс 1
mask_train1 = y_train == 1
plt.scatter(x_train[mask_train1, 0], x_train[mask_train1, 1], c='blue', marker='x', label='Train Class 1')
    
# Тестовые точки - правильно классифицированные
correct = y_test == y_pred
mask_test0_correct = (y_test == 0) & correct
mask_test1_correct = (y_test == 1) & correct
    
plt.scatter(x_test[mask_test0_correct, 0], x_test[mask_test0_correct, 1], 
                c='green', marker='o', label='Test Class 0 Correct')
plt.scatter(x_test[mask_test1_correct, 0], x_test[mask_test1_correct, 1], 
                c='green', marker='x', label='Test Class 1 Correct')
    
# Тестовые точки - неправильно классифицированные
incorrect = y_test != y_pred
mask_test0_incorrect = (y_test == 0) & incorrect
mask_test1_incorrect = (y_test == 1) & incorrect
    
plt.scatter(x_test[mask_test0_incorrect, 0], x_test[mask_test0_incorrect, 1], 
            c='red', marker='o', label='Test Class 0 Incorrect')
plt.scatter(x_test[mask_test1_incorrect, 0], x_test[mask_test1_incorrect, 1], 
            c='red', marker='x', label='Test Class 1 Incorrect')
    
plt.show()
```
